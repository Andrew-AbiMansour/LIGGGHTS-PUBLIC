BACKGROUND
==========
The Beverloo equation relates the mass flow rate, $\dot{W}$, of granular materials to the mean particle diameter, $d$. The equation assumes $\dot{W}$ is constant in time for non-cohesive materials, the form of which is

$$ \dot{W} = S (D_0 - k d)^{\frac{5}{2}}, $$ where $S$ is a time-independent scaling parameter 

$$ S = C \rho_b \sqrt{g}. $$

The beverloo constant $k$ is related to the shape of the particle, while $D_0$ is the inlet diameter. The bulk density $\rho_b$ is computed from DEM simulations, while the dimensionless parameter $C$ is taken to be 0.58. The objective of this study is to validate the Beverloo equation for non-cohesive powders using DEM simulation. In the first part of this study, the power dependence of $\dot{W}$ on $d$ is investigated, and the beverloo constant $k$ is then deduced and compared to its well known empirical value of $1.4$ for spherical particles. In the second part, the Beverloo correlation is shown to break down when the particle diameter is larger than $1000$ $\mu m$ or when the height of granular flow is close to the outlet diameter, $D_0$.

MODEL
=====
We consider a flat-bottomed hopper The contact model used is Hooke's law:

$$\mathbf{f}_{ij} = \mathbf{f}_{ij}^n + \mathbf{f}^t_{ij},$$

where the normal force, $\mathbf{f}_{ij}^n$, takes the following expression:

$$\mathbf{f}_{ij}^n = k_n \delta \mathbf{n}_{ij} - \gamma_n \delta \mathbf{\dot{n}}_{ij},$$

and similarly the tangential force $\mathbf{f}^t_{ij}$ is

$$\mathbf{f}_{ij}^t = k_t \delta \mathbf{t}_{ij} - \gamma_t \delta \mathbf{\dot{t}}_{ij}$$

METHODOLOGY
===========
The time-averaged mass flow rate is computed computationally from DEM simulations for various particle sizes, and then it is compared to the flow rate obtained from the Beverloo equation. In a rectangular box of dimensions $3.2$ cm $\times$ $3.2$ cm $\times$ $6.2$ cm, a stopped (wall) at $z=0$ is placed along the $xoy$ plane. A number of spherical particles ($N = 428,325 - $) of constant diameter ($d = 400 - 1,800$ $\mu m$) is inserted in a cylindrical region for a total of $10^5$ steps, each of length $\delta t = 20$ $\mu s$. Rolling friction is set to zero.

<h2>THE MASS FLOW RATE</h2>

In this section, we compute the flow rate from DEM simulations and compare it to that predicted from the Beverloo equation for particle diameter $d = (250-600)$ $\mu m$. We assume that both the Beverloo constant, $k$, and the exponent $\alpha$ are unknown, which are determined from a series of $8$ different simulations, each corresponding to a different particle size.

<h2>DATA</h2>

```python
from numpy import array, sqrt, log, loadtxt, ones, arange, \
                  exp, polyfit, arange

# Some global constants and parameters
g = 9.81 # gravitational acceleration in m/s^2
rho_b = 1600.0 # bulk density (Kg/m^3) computed from DEM, 
            # seems to very very little with particle diameter
C = 0.58 # empirical constant [not quite sure what this means]
scaling = sqrt(g) * rho_b * C
D0 = 1e-2 # Hopper outlet diameter in meters
k = 1.4 # Beverloo constant for spherical particles
alpha = 2.5

def Beverloo(d):
	# units in Kg/s
	return scaling * (D0 - k * d)**alpha

# Particle size in meters
d = array([200.0, 250.0, 300.0, 350.0, 400.0, 450.0, 500.0, 550.0, \
           600.0, 700.0, 800.0, 900.0, 1000.0, 1050.0]) * 2.0 * 1e-6

# computed from the Beverloo function above for units in Kg/s
rateBev = Beverloo(d)

# computed as a time average from DEM simulations
rateDEM = array([0.026413673273341997, 0.024922974469244324, 0.023767808980523674, \
		 0.022541457777739922, 0.021628972546264197, 0.020619523022059421, \
		 0.019663200860988245, 0.018587169666778633, 0.017637529475783811, \
		 0.016390980417424414, 0.015320919053026716, 0.014227998238750979, \
		 0.013738517873369011, 0.013501568751209125]) 
```

<h3>Logarithmic method</h3>

Taking the natural logarithm of the mass rate in the Beverloo equation yields:

$$ \log{\dot{W}} = \log{(C\rho_b \sqrt{g})} + \alpha \log{(D_0 - k d)}.$$

Therefore, a plot of log of the mass rate, $\dot{W}$, versus $\log{(D_0 - k d)}$, should yield a straight line of slope $\alpha$. From the y-intercept, one can then determine the value of $C$. The slope seems inaccurate while the y-intercept is close to its theoretical value. The reason is the the variation in the mass rate is relatively small across the small interval of particle diameters simulated here. Taking the logarithm of the mass rate, $W$, seems to amplify the error in the intercept. The result is an inaccurate way (rel. error more than  100%) of determining the $C$ constant in the Beverloo equation.

```python
# First method based on log plots ~ not quite accurate enough
plt.plot(log(D0 - k*d), log(rateDEM), '-*')
plt.plot(log(D0 - k*d), log(rateBev), '-.')
plt.xlabel('$\log{(D_0 - kd)}$'); plt.ylabel('$\log{\dot{W}}$'); 
plt.legend(['DEM', 'Beverloo'], loc='best'); plt.grid()

# Exclude 400e-6 < d < 6 * D0 points
nPts = arange(1,len(d) - 3)

# Do linear regression
slope, inter = polyfit(log(D0 - k * d[nPts]), log(rateDEM[nPts]), 1)
alpha = slope
C = exp(inter) / (rho_b * sqrt(g))
print 'C = ', C, ', alpha = ', alpha
```

<h3>Power expansion method</h3>

Provided $d/D_0 \ll 1$, we can performing a 1st order Maclaurin expansion for $(1 - k d /D_0)$ in terms of $d/D_0$, the Beverloo equation then becomes:


$$ \frac{W}{S} = C \left[ 1 - \alpha k \left(\frac{d}{D_0} \right) \right] + O\left( \left[\frac{d}{D_0}\right]^2 \right),$$ where

$$ S = \rho_b \sqrt{g} D_0^{\alpha}. $$

In this case, a plot of the scaled mass rate, $W/S$, versus the particle diameter, $d$, yields a straight line with higher accuracy. The noise (due to numerical/simulation error) is not amplified, which leads to much more accurate results (less than 3.0 % rel. error) in both $C$ and $\alpha$. . 

```python
# Second method based on power expansion ~ better accuracy
alpha = 2.5
scaling = rho_b * sqrt(g) * D0**alpha

plt.figure()
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.plot(d / D0, rateDEM / scaling, '-*')
plt.plot(d / D0, rateBev / scaling, '-.')
plt.xlabel('$d/D_0$'); plt.ylabel('$\dot{W}/S$'); 
plt.legend(['DEM', 'Beverloo']); plt.grid()
slope, inter = polyfit(d[nPts] / D0, rateDEM[nPts]/scaling, 1)
C = inter
alpha = - slope / (k * C)

print 'C = ', C, ', alpha = ', alpha
```

```python
# Particle size in meters
d = array([200.0, 250.0, 300.0, 350.0, 400.0, 450.0, 500.0, 550.0, 600.0, \
           700.0, 800.0, 900.0, 1000.0, 1050.0]) * 2.0 * 1e-6

# computed from the Beverloo function above for units in Kg/s
rateBev = Beverloo(d)

# computed as a time average from DEM simulations
rateDEM = array([0.026413673273341997, 0.024922974469244324, 0.023767808980523674, \
		 0.022541457777739922, 0.021628972546264197, 0.020619523022059421, \
		 0.019663200860988245, 0.018587169666778633, 0.017637529475783811, \
		 0.016390980417424414, 0.015320919053026716, 0.014227998238750979, \
		 0.013738517873369011, 0.013501568751209125]) 

# compute the % rel. error
error = abs((rateBev - rateDEM)) / rateBev * 100.0

plt.plot(d / D0, error)
plt.xlabel('$d/D_0$'); plt.ylabel('% rel. error')
plt.grid()
```

<h2>BREAKDOWN IN THE BEVERLOO EQUATION</h2>

In this section we study the spatial regimes in which the Beverloo equation breaks down. If we introduce the height function $H(t)$ that specifies the height of the powder at time $t$:

$$H(t) = \max{ \{z_i | i \in [1,N] \} }$$

The Beverloo equation assumes the flow rate is independent of $H(t)$, which seems to be true only when $H(t) \geq 2D_0$. This is shown in the figure below for a system comprising monodispersed particles of radius $200$ $\mu m$. The scaled mass flow rate $\dot{W}/D_0$ and the ratio $H/D_0$ are plotted as a function of time. At $t=[0.3-1.0]$ s, the system is at equilibrium, and $H(t)$ is approxiomately a constant. At $t > 1$ s, the outlet is removed, and the powder begins to flow with approximately a constant rate, while $H$ decreases in time. When $H < 2D_0$, however, the flow rate begins to collapse until it completely vanishes. 

```python
ax = plt.subplot(111)

# Load data files for monodispersed (200 microns) sim
height = loadtxt(dataDir+'/Monodispersed/out-radius-200/height.dat')
flow = loadtxt(dataDir+'/Monodispersed/out-radius-200/flow.dat')
time = arange(0,len(height)) * dt * freq

plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

plt.plot(time, height[:] / D0);
plt.plot(time, flow[:] / D0);
plt.plot(time, ones(len(flow[:])) * 2.0, '--');

ax.annotate('outlet removed', xy=(101*dt*freq, 0.02), xytext=(110*dt*freq, -2.1), \
            arrowprops=dict(facecolor='black', shrink=1.0))
ax.annotate('outlet removed', xy=(101*dt*freq, 3.3), xytext=(110*dt*freq, 5.2), \
            arrowprops=dict(facecolor='black', shrink=1.0))

ax.legend(['$H/D_0$', '$\dot{W}/D_0$'], loc='lower right')
plt.xlabel('Time (s)')
plt.grid()
```
